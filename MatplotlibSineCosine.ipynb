```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python

mylist1 = np.arange(-60, 360, 30)
sine = np.sin(mylist1)
cosine = np.cos(mylist1)
plt.figure(figsize=(10,6))
plt.plot(mylist1,sine,linewidth='3',label='sine')
plt.plot(mylist1,cosine,linewidth='3',color='yellow',label='cosine')
plt.title('Sine-Cosine Graph')
plt.xlabel('Sine & Cosine')
plt.ylabel('Values')
plt.legend()

plt.show()
```
